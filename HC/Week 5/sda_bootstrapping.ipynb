# SDA: bootstrapping

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams["figure.figsize"] = (6, 3)
plt.rcParams["xtick.minor.visible"] = True
plt.rcParams["ytick.minor.visible"] = True
```

```python
np.random.seed(10)
bins = 10
```

**Step 1**: Simulate a data set of 20 measurements, plot the distribution, and compute the sample mean.

```python
data = np.random.normal(loc=5, scale=2, size=20)
# data = np.random.exponential(scale=2, size=20)
# data = np.random.lognormal(mean=5, sigma=2, size=20)
```

```python
# Create your plot here
plt.hist(data, bins=10, edgecolor="black")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Random Data")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```

```python
# Calculate the sample mean here
mean = np.mean(data)
print(f"Sample mean: {mean:.2f}")
```

**Step 2**: Generate 1000 bootstrap samples, compute their means, and plot the distribution.

```python
n_boot = 1000
boot_mean = np.zeros(n_boot)

for i in range(n_boot):
    # Set the arguments in the choice function
    boot_sample = np.random.choice(data, size=len(data))
    boot_mean[i] = np.mean(boot_sample)
```

```python
# Create your plot here
plt.hist(boot_mean, bins=bins, edgecolor="black")
plt.xlabel("Bootstrap Mean")
plt.ylabel("Frequency")
plt.title("Distribution of Bootstrap Means")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```

**Step 3**: Estimate the 68% confidence interval for the mean of $x$ by taking the 16th and 84th percentiles of the bootstrap distribution. Plot again the distribution of the bootstrap means and show the confidence interval (e.g. with vertical lines).

```python
# Calculate the percentiles here
percent = np.percentile(boot_mean, [16, 84])
```

```python
# Create your plot here
plt.hist(boot_mean, bins=bins, edgecolor="black")
plt.xlabel("Bootstrap Mean")
plt.ylabel("Frequency")
plt.axvline(percent[0], color="red", linestyle="--", label="16th Percentile")
plt.axvline(percent[1], color="green", linestyle="--", label="84th Percentile")
plt.axvline(mean, color="blue", linestyle="--", label="Sample Mean")
plt.legend()
plt.title("Distribution of Bootstrap Means")
plt.grid(True, which="both")
plt.show()
```
